```python
def all_zeros_success(outcome):
    """Success criteria: all measurement outcomes must be 0."""
    clean = outcome.replace(" ", "")
    return all(bit == "0" for bit in clean)


executor = QuantumCircuitExecutor(save_statevector=True, shots=2048)
# results = executor.simulate(generator.circuit, success_criteria=all_zeros_success, show_results=True)

results_pauli = executor.simulate(
    generator.circuit,
    success_criteria=all_zeros_success,
    show_results=True,
    noise_model="pauli",
    noise_level=0.1,
)
```
